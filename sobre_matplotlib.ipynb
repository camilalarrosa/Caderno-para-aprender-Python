```python
# Biblioteca para modelagem de dados
import pandas as pd
 
# Biblioteca para recursos matemáticos
import numpy as np

# Biblioteca para recursos Graficos (visualização de dados)
import matplotlib.pyplot as plt
```

### Gráfico de linhas

```python
np.random.seed(7)

# gerando vetor com nros aleatorios entre 1 e 15000, com 10 amostras

y=np.random.randint(low=1, high= 1500, size=10)
```

```python
y
```

```python
plt.plot(y)
```

```python
# insere a primeira linha no plot (linha tracedaja)
plt.plot(y,color= '#749187', marker = 'o', ms = 5, mec = 'k',
         markerfacecolor = 'w', ls = '-.')

      # y,color (cor da linha), marker(tipo de marcador), 
      # ms(tamanho do marcador), mec(cor da linha do marcador),
      # markerfacecolor (cor da face do marcador) & ls(estilo da linha);

# insere a segunda linha no plot (valores x 2)
plt.plot(y*2, marker = '+', color = 'k', ms = 5)

#rotulos
plt.xlabel('Eixo X', color = 'red', size = 12)
plt.ylabel('Eixo Y')
plt.title('Título', loc = 'left')

#gridlines
plt.grid(axis = 'y', color = 'gray',
         linestyle = '--', linewidth = 1,
         alpha = 0.8)
      # obs: alpha = opacidade da linha
plt.show()
```

### Subplots

```python
# criando as funcoes para as figuras
np.random.seed(6)
x = np.arange(1,11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

#colocando o titulo geral no dashboard
plt.figure(figsize = (15,5))
plt.suptitle('Figura', fontsize = 15)

plt.subplot(1,3,1) #quero que o subplot tenha 1 linha e 3 colunas, e o q eu vou
                   # inserir aqui, vai estar na 1ª entrada --> linha 1, coluna 1
plt.plot(x, y1, color = 'k')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1,3,2)
plt.plot(x, y2, color = 'r')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1,3,3)
plt.plot(x, y2, color = 'g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.tight_layout(pad=4) #aumentando espaço entre graficos
plt.show()

```

```python
# outra maneira 
fig, ax = plt.subplots(1, 3, figsize = (15,5)) # 1 linha, 3 colunas, tamanho da fig 15 x 5
fig.suptitle('Figure')
ax[0].plot(x, y1, color = 'black') #linha 0, coluna 0
ax[1].plot(x, y2, color = 'red') #linha 0, coluna 1
ax[2].plot(x, y3, color = 'green') # linha 0, coluna 2

for i in range(3):
  ax[i].set(title = f'Subplot {i+1', xlabel = 'Eixo X', ylabel = 'Eixo Y')
```

### Scatter plot

```python
# grafico muito utilizado para ver a relacao entre variaveis
x = np.random.normal(10,0.5,100)
y = np.random.uniform(0,20,100)

fig = plt.figure()
ax = plt.axes()
ax.scatter(x, y, marker = 'o', color = 'red', label = 'data 1', alpha = 0.5)
ax.scatter(x*0.5, y*0.5, marker = 'v', color = 'black', label = 'data 2', alpha = 0.7)
ax.legend();
# obs: label = legenda, alpha = opacidade
```

### Dados econômicos

```python
import plotly.express as px
```

```python
px.data.gapminder()
```

```python
df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
df.head()
```

```python
# grafico evolução renda per capita no BR
plt.plot(df.index, df['gdpPercap']) #eixo x = index(ano), eixo y = renda per capita
plt.title('PIB per capita do Brasil')
plt.ylabel('PIB per capita')
plt.xlabel('Tempo');
```

```python
# grafico expectativa de vida x renda per capita

title = 'Relação entre expectativa de vida e renda per capita no Brasil'
plt.figure(figsize = (12,4))
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap = 'viridis') #cmap = colormap
plt.xlabel('Expectativa de vida')
plt.ylabel('Renda per capita')
plt.title(title, loc = 'left')
plt.show()
```

### Gráfico de barra

```python
# eixo x = index (ano), altura da barra (height) = população
plt.bar(x= df.index,
        height = df['pop'],
        color = 'red')
plt.title('População brasileira');
```

```python
# função filtrar o dataset pelo continente
def filtrar_continente(continente):
  df = px.data.gapminder()
  df = df[df['continent'] == continente]
  return df
```

```python
# função filtrar o pais do dataset por uma variavel
def filtrar_pais(pais, variavel):
  'Filtra algum país da amostra para as variaveis "pop", "gdpPercap" e "lifeExp"'
  df = px.data.gapminder()
  df = df[df['country'] == pais][variavel]
  return df
```

```python
americas = filtrar_continente('Americas')  #1º filtrei o continente q eu quero
paises = americas['country'].unique()
plt.figure(figsize = (12, 8))

#pra cada pais dentro da lista de paises, vou fazer um scatterplot diferente
for pais in paises:
  plt.scatter(filtrar_pais(pais = pais,
                           variavel = 'lifeExp'),
              filtrar_pais(pais = pais,
                           variavel = 'gdpPercap'))
  plt.legend(labels = paises, loc = 'best')
  plt.title('Relação entre renda Per capita e expectativa de vida', loc = 'left')
  plt.xlabel('Expectativa de vida')
  plt.ylabel('Renda per capita')
```
